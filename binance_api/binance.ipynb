```python
# 분봉 데이터 조회

import ccxt
import pandas as pd 

binance = ccxt.binance()
btc_ohlcv = binance.fetch_ohlcv("BTC/USDT") # 500건 조회

df = pd.DataFrame(btc_ohlcv, columns=['datetime', 'open', 'high', 'low', 'close', 'volume'])
df['datetime'] = pd.to_datetime(df['datetime'], unit='ms') # 밀리세컨드 단위의 타임스탬프 형태 데이터를 datetime으로 설정
df.set_index('datetime', inplace=True) # 시계열 데이터 >> 인덱스를 시간 값으로 설정, inplace값을 T로 준 것은 원본(df)에 바로 적용
print(df) # UTC 기준이기 때문에 한국시간보다 9시간 느림
```

```python
# 일봉 데이터 조회 

import ccxt 
import pandas as pd 


binance = ccxt.binance()
btc_ohlcv = binance.fetch_ohlcv("BTC/USDT", timeframe = '1d') # timeframe 생략하고 "1d"만 줘도 됨

df = pd.DataFrame(btc_ohlcv, columns=['datetime', 'open', 'high', 'low', 'close', 'volume'])
df['datetime'] = pd.to_datetime(df['datetime'], unit='ms')
df.set_index('datetime', inplace=True)
print(df)
```
